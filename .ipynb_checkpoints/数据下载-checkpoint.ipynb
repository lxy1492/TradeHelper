```python
import json
import socket
from Moudle.dataType import DataType
import pickle
```

```python
def downloaddata(name,date,ip="111.230.177.71",port=8001,batchsize=10,recvSize=1024000):
    c = socket.socket(socket.AF_INET,socket.SOCK_STREAM)
    c.connect((ip,port))
    print(c.recv(1024).decode())
    mess = "downloaddata "+name+" "+date+" "+str(batchsize)
    c.send(mess.encode())
    r = c.recv(recvSize)
    data = r.decode("gbk")
    datas = []
#     print(data)
    data = json.loads(data)
    if data["result"]=="success":
        datas.extend(data["data"])
    else:
#         print(data)
        return -1
    while(True):
        c.send(b"next data")
        data = c.recv(recvSize)
        data = json.loads(data.decode("gbk"))
        if data["result"]=="success":
            datas.extend(data["data"])
            print(data["info"],data["name"],data["date"])
        else:
            print(data)
            break
    return datas
```

```python
def getList(ip="111.230.177.71",port=8001,recvSize=102400):
    c = socket.socket(socket.AF_INET,socket.SOCK_STREAM)
    c.connect((ip,port))
    print(c.recv(1024).decode())
    mess = "databaselist"
    c.send(mess.encode())
    r = c.recv(recvSize)
    return json.loads(r.decode())
```

```python
l = getList(ip="111.230.177.71",port=8001)
```

```python
for each in l:
    r = downloaddata(each,"",batchsize=40,recvSize=102400)
    with open("../FileWebDIR/TradeHelperDataBase/"+each,"w") as f:
        for each in r:
            f.write(json.dumps(each)+"\n")
```
