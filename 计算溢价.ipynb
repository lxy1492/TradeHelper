```python
DataBaseDIR = "../FileWebDIR/TradeHelperDataBase/Original/"
```

```python
fileList = {}
for each in os.listdir(DataBaseDIR):
    if each.split(".")[-1]=="tbs":
        name = each.split(" ")[0]
        if name in fileList:
            fileList[name].append(each)
        else:
            fileList.update({name:[each]})
```

```python
datas = []
with open("./Ag(T+D).tbs","r") as f:
    for each in f.read().split("\n"):
        if each!="":
            base64_data = base64.b64decode(each)
            datas.append(pickle.loads(base64_data))
```
