```python
from Moudle.dataType import DataType
from manager.dataBaseManager import getALLOrigianlList,readData,getOriginalData
from matplotlib import pyplot as plt
```

```python
fileList = getALLOrigianlList()
```

```python
datas = getOriginalData(fileList["Ag(T+D)"])
```

```python
premiumDatas = []

for each in datas:
    if "premium" in each.info:
        premiumDatas.append(each)
```

```python
plt.figure(figsize=(16,9),dpi=200)
plt.plot([t.timeStamp for t in premiumDatas],[p.info["premium"] for p in premiumDatas])
plt.show()
```
